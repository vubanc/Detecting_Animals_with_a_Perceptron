# Training and testing the perceptron

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
#def sigmoid_derivative(x):
    #return x*(1-x)
```

```python
def predict(w,x):
    return sigmoid(np.dot(w,x.T))

def loss(y_bar, y):
    n=len(y)
    #Calculating total_loss as max(0, -y*f(x))/n
    total_loss = -y_bar*y
    total_loss[total_loss < 0] = 0
    return np.sum(total_loss)/n
```
